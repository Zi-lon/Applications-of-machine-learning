 # How to load a csv dataset

```python
columns = [
    "class",
    "Alcohol",
    "Malic acid",
    "Ash",
    "Alcalinity",
    "Magnesium",
    "Total phenols",
    "Flavanoids",
    "Nonflavanoid phenols",
    "Proanthocyanins",
    "Color intensity",
    "Hue",
    "OD280/OD315 of diluted wines",
    "Proline",
]
data = pd.read_csv('./wine.data', names=columns)
# data = sklearn.datasets.load_wine(names=columns)
data
```

 # How to do data exploration


```python
X, y = data.iloc[:, 1:], data["class"]
display(y)
display(X)
```

```python
# Normalize
X = (X - X.min())/(X.max()- X.min())
X.describe()
```

 # How to do classification
 ## Train - Test split

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.3)
print(f"Shape of training features: {train_X.shape}")
print(f"Shape of validation features: {val_X.shape}")
```

```python
# ## Train a classifier model
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_obj = KNeighborsClassifier(n_neighbors=3)
classifier_obj.fit(train_X, train_y)
```

 ## How to use an sklearn model

```python
from sklearn.metrics import accuracy_score
predicted_train_y = classifier_obj.predict(train_X)
predicted_val_y = classifier_obj.predict(val_X)
acc_train = accuracy_score(train_y, predicted_train_y)
acc_val = accuracy_score(val_y, predicted_val_y)
print(f"Train accuracy is {acc_train:.2f} and val accuracy is {acc_val:.2f}")
```
